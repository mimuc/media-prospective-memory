````r
library(dplyr)
library(report)
source('stats.R')
````

````r
df_ddm <- read.csv('data/ddm.csv')
head(df_ddm)

df_ddm_pm = df_ddm %>% filter((task %in% c('PM')))
head(df_ddm_pm)

df_ddm_ld = df_ddm %>% filter((task %in% c('LD')))
head(df_ddm_ld)
````

````r
df_ddm_pm$interrupt <- factor(df_ddm_pm$interrupt, levels=c('rest', 'twitter', 'youtube', 'tiktok'))
df_ddm_pm$measure   <- factor(df_ddm_pm$measure, levels=c('pre', 'post'))

df_ddm_ld$interrupt <- factor(df_ddm_ld$interrupt, levels=c('rest', 'twitter', 'youtube', 'tiktok'))
df_ddm_ld$measure   <- factor(df_ddm_ld$measure, levels=c('pre', 'post'))
````

````r
twoway_anova_test(df_ddm_pm, 'drift')
````

````r
twoway_anova_test(df_ddm_pm, 'noise')
````

````r
twoway_anova_test(df_ddm_pm, 'bound')
````

````r
twoway_anova_test(df_ddm_pm, 'nondectime')
````

````r
# twoway_anova_test(df_ddm_ld, 'drift')
m <- aov(drift ~ interrupt*measure, data=df_ddm_ld)
summary(m)
````

````r
eta_squared(m, partial = TRUE)
````

````r
twoway_anova_test(df_ddm_ld, 'noise')
````

````r
twoway_anova_test(df_ddm_ld, 'bound')
````

````r
twoway_anova_test(df_ddm_ld, 'nondectime')
````

````r
library(lme4)
library(dplyr)
library(bayestestR)
library(report)
````

````r
df <- read.csv("data/acc.csv")
````

# LD

````r
df_ld <- df %>% filter(task == "LD")
df_ld
````

````r
# LMM on the main effect `interrupt` w/wo link family
m0 <-  lmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld)
m1 <- glmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld, family = Gamma(link = "log"))
m2 <- glmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld, family = Gamma(link = "inverse"))
m3 <- glmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "log"))
m4 <- glmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "inverse"))
m5 <- glmer(accuracy  ~ interrupt + (1|folder_id), data = df_ld, family = binomial(link = "logit")) # suggested by R2

# LMM on the main effect `measure` w/wo link family
m6  <-  lmer(accuracy  ~ measure + (1|folder_id), data = df_ld)
m7  <- glmer(accuracy  ~ measure + (1|folder_id), data = df_ld, family = Gamma(link = "log"))
m8  <- glmer(accuracy  ~ measure + (1|folder_id), data = df_ld, family = Gamma(link = "inverse"))
m9  <- glmer(accuracy  ~ measure + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "log"))
m10 <- glmer(accuracy  ~ measure + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "inverse"))
m11 <- glmer(accuracy  ~ measure + (1|folder_id), data = df_ld, family = binomial(link = "logit")) # suggested by R2

# LMM on the interaction `interrupt * measure` w/wo link family
m12 <-  lmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld)
m13 <- glmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld, family = Gamma(link = "log"))
m14 <- glmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld, family = Gamma(link = "inverse"))
m15 <- glmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "log"))
m16 <- glmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld, family = inverse.gaussian(link = "inverse"))
m17 <- glmer(accuracy ~ interrupt * measure + interrupt + measure + (1|folder_id), data = df_ld, family = binomial(link = "logit")) # suggested by R2
````

````r
anova(m0, m1, m2, m3, m4, m5, m6, m7, m8, m9, m10, m11, m12, m13, m14, m15, m16, m17)
````

````r
bayesfactor_models(m1, m2, m3, m4, m5, m6, m7, m8, m9, m10, m11, m12, m13, m14, m15, m16, m17, denominator = m0)
````

````r
df_ld$interrupt <- factor(df_ld$interrupt, levels = c("tiktok", "rest", "twitter", "youtube"))
df_ld$measure   <- factor(df_ld$measure, levels = c("post", "pre"))
m <- lmer(accuracy ~ interrupt + (1 | folder_id), data = df_ld)
print(report(m), verbose = TRUE)
````

````r
df_ld$interrupt <- factor(df_ld$interrupt, levels = c("rest", "twitter", "youtube", "tiktok"))
df_ld$measure   <- factor(df_ld$measure, levels = c( "pre", "post"))
m <- lmer(accuracy ~ interrupt + (1 | folder_id), data = df_ld)
print(report(m), verbose = TRUE)
````

````r
df_ld$interrupt <- factor(df_ld$interrupt, levels = c("tiktok", "rest", "twitter", "youtube"))
df_ld$measure   <- factor(df_ld$measure, levels = c("post", "pre"))
m <- lmer(accuracy ~ measure + (1 | folder_id), data = df_ld)
print(report(m), verbose = TRUE)
````

````r
df_ld$interrupt <- factor(df_ld$interrupt, levels = c("rest", "twitter", "youtube", "tiktok"))
df_ld$measure   <- factor(df_ld$measure, levels = c( "pre", "post"))
m <- lmer(accuracy ~ measure + (1 | folder_id), data = df_ld)
print(report(m), verbose = TRUE)
````
